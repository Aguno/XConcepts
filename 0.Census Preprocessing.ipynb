```python
# Load libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import os

url = "./adult.csv"
df = pd.read_csv(url)
```

```python
#Process Nans
col_names = df.columns
for c in col_names:
	df[c] = df[c].replace("?", np.NaN)

df = df.apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
#replace marital status to divorced, married, not married
df.replace(['Divorced', 'Married-AF-spouse', 
              'Married-civ-spouse', 'Married-spouse-absent', 
              'Never-married','Separated','Widowed','<=50K','>50K'],
             ['divorced','married','married','married',
              'not married','not married','not married',0,1], inplace = True)
df.head(10)
```

```python
df=df.drop(['fnlwgt','education','native-country'],axis=1)
```

```python
category_col_1 =['workclass', 'marital-status','race','gender', 'occupation',
               'relationship'] 

df = pd.get_dummies(df, columns=category_col_1, drop_first=True)
```

```python
df.describe()
```

```python

df =df[[c for c in df if c not in ['income']] + ['income']]

for c in df.columns:
    df[c] = df[c].astype(int)
df.head(20)
if not os.path.exists("metadata"):
    os.makedirs("metadata")
df.to_csv('metadata/Census_income.csv',index=False)
```
